```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("C:\\Users\\RAHUL\\Desktop\\csvfolder\\HR_EmployeeAttrition.csv")
df.head()
```

```python
def new_Attrition(X):
    if X == 'Yes':
        return 1
    if X == 'No':
        return 0
```

```python
df['Attrition']=df['Attrition'].apply(new_Attrition)
df.head()
```

```python
attrition_there=df[df['Attrition']==1]
attrition_there.shape
```

```python
attrition_notthere=df[df['Attrition']==0]
attrition_notthere.shape
```

```python
df.groupby('Attrition').mean()
```

```python
dfnew=df[['Attrition','DistanceFromHome','EnvironmentSatisfaction','JobInvolvement','JobLevel','StockOptionLevel','TotalWorkingYears','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
dfnew.head()
```

```python
dummies=pd.get_dummies(dfnew['Attrition'],prefix='Attrition')
dummies
```

```python
merger=pd.concat([dfnew,dummies],axis='columns')
merger
```

```python
final=merger.drop(['Attrition'],axis='columns')
final
```

```python
X=final
X.head()
```

```python
y=df['Attrition']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(X_train,y_train) # train the model
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
